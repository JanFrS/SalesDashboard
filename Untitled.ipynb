# Sales report year 2022

This is the project realized to create a dashboard of the sales made in the year 2022 by a diesel engines workshop, the idea of the project is to improve the last one created and include more dynamic data visualization, in the current notebook you will find the steps taken to achieve the final dashboard

```python
#Import the libraries 
import numpy as np
import pandas as pd
import panel as pn
import xlrd
pn.extension('tabulator')

import hvplot.pandas
```

```python
#Read and extrac the data
df1 = pd.read_excel('Sales.xls',index_col=1)
df1
```

## Clients order by purchases
The first step we are going to do is to group the clients per month and organize them depending on the total of the invoices paid, the dashboard will include a slider where you can change the month to see the different clients and the total value paid

```python
salesdf = df1.groupby ([df1['FECHA'].dt.month,'TERCERO'])['SUBTOTAL'].sum().reset_index(name='TOTAL')
salesdf.rename (columns = {"TERCERO":"CLIENT","FECHA":"Month"},inplace = True)
isalesdf = salesdf.interactive() 
```

```python
month_slider = pn.widgets.IntSlider (name = 'Month', start = 1, end = 12, step = 1, value = 2)
```

```python
clients_pipeline = (
    isalesdf [
        (isalesdf.Month == month_slider)].sort_values(by = 'TOTAL', ascending = False))
clients_tb = clients_pipeline.pipe (pn.widgets.Tabulator,pagination = 'remote', page_size = 10,sizing_mode = 'stretch_width')
clients_tb
```

```python
clients_plot = clients_pipeline.hvplot.barh('CLIENT','TOTAL',stacked = True, height = 200,title ="Clients ordered by purchases")
clients_plot
```

```python
# Checkbox group creation
checkbox_group = pn.widgets.CheckBoxGroup(
    name='Checkbox Group', value=['Month','Subtotal', 'Taxes', 'Total'], options=['Month','Subtotal', 'Taxes', 'Total'],
    inline=True)
```

```python
sales_by_monthdf = df1.groupby(df1['FECHA'].dt.month)[['SUBTOTAL','IVA','TOTAL']].agg("sum").reset_index(names='Month')
sales_by_monthdf.rename (columns = {"SUBTOTAL":"Subtotal","IVA":"Taxes","TOTAL":"Total"},inplace = True)
isales_by_month = sales_by_monthdf.interactive()
```

```python
sales_by_month_pipeline = isales_by_month.loc[:,checkbox_group]
```

```python
sales_plot = sales_by_month_pipeline.hvplot.bar (x='Month',stacked=False, rot = 90,title ="Values billed within the year")
sales_plot
```

```python
df2 = pd.read_excel('Sales.xls','COSTOS Y GASTOS',index_col=1)
```

```python
purchasesdf=df2.groupby(df2['FECHA'].dt.month)['TOTAL A PAGAR'].sum().reset_index(name='Total')
```

```python
salesdf2 = df1.groupby(df1['FECHA'].dt.month)['TOTAL'].sum().reset_index(name='Total')
```

```python
monthly_income = pd.merge (purchasesdf,salesdf2,on='FECHA')
monthly_income['Income'] = monthly_income['Total_x'] - monthly_income['Total_y']
monthly_income.rename (columns = {"FECHA":"Month"},inplace = True)
```

```python
income_plot = monthly_income.hvplot(x='Month',y='Income',width = 400,title = "Income per month during the year")
income_plot
```

```python
template=pn.template.FastListTemplate(
    title = 'Year sales report dashboard',
    sidebar = [pn.pane.Markdown("# Year sales report"),
             pn.pane.Markdown("##Sales report year 2022"),
             pn.pane.Markdown("###This project was created to show a dashboard of the data taken in a diesel engines workshop, where the data with the sales info is stored in an Excel document. The idea is to give them a yearly report about their sales, the total of the sales per year, per month, and categorization of the clients by the purchases during the year")],
    main = [pn.Row(pn.Column(income_plot),(sales_plot)),
           pn.Row(pn.Column(month_slider,clients_plot.panel(width=600)),(clients_tb))]
)
template.show()
```
